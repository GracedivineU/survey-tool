```python
import pandas as pd

df = pd.read_csv(r"C:\Users\user\Desktop\Survey_tool_project\survey_tool\user_data.csv")
df.head()
```

```python
# the gender distribution across spending categories.
expense_categories = ['Utilities', 'Entertainment', 'School Fees', 'Shopping', 'Healthcare']

df_melted = df.melt(id_vars=['Gender'], value_vars=expense_categories,
                    var_name='Category', value_name='Amount')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='Category', y='Amount', hue='Gender')
plt.title("Gender-wise Spending by Category")
plt.ylabel("Average Amount Spent")
plt.tight_layout()
plt.savefig("chart_gender_spending.png")  # Saving chart for PowerPoint
plt.savefig(r"C:\Users\user\Desktop\Survey_tool_project\survey_tool\chart_gender_spending.png")
plt.show()
```
